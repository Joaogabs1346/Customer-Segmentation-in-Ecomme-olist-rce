### Data processing

```python
customer_orders = olist_customers.merge(olist_orders, how='inner',on='customer_id')
customer_orders =customer_orders.merge(olist_order_items, how='inner',on='order_id')
```

```python
display(customer_orders)
```

```python
customer_orders.columns
```

```python
customer_orders.drop(columns=[
    i for i in customer_orders.columns if i not in['customer_unique_id','order_id','order_status','price','freight_value']
],inplace= True )
```

```python
customer_orders.order_status.unique()
```

```python
customer_orders = customer_orders[(customer_orders.order_status != 'unavailable') & (customer_orders.order_status != 'canceled')]
```

```python
customer_orders
```

####  Total Amount spended by customers


```python
values_by_customer= customer_orders.groupby(by ='customer_unique_id').sum()
```

```python
values_by_customer['Total_value'] = values_by_customer['price'] + values_by_customer['freight_value']
```

```python
values_by_customer = values_by_customer.reset_index(level=0)
```

```python
values_by_customer
```

```python
values_by_customer.describe()
```

#### Number of orders placed

```python
customer_orders
```

```python
customer_orders.order_status.unique()
```

```python
customer_orders = customer_orders[(customer_orders.order_status != 'unavailable') & (customer_orders.order_status != 'canceled') & (customer_orders.order_status != 'created')]
```

```python
qnt_orders_by_cust = customer_orders.groupby(by='customer_unique_id').count()['order_id']
```

```python
qnt_orders_by_cust = qnt_orders_by_cust.reset_index(level=0)
```

#### Creating the column with the requested qnt and the total amount

```python
customer_info = values_by_customer.merge(qnt_orders_by_cust, how= 'inner', on='customer_unique_id')
```

```python
customer_info
```

```python
customer_info.drop(columns=['price','freight_value'],inplace=True)
customer_info.rename(columns={'order_id':'qnt_orders'},inplace=True)
```

```python
customer_info.describe()
```

### Machine learning

```python
features = customer_info[['Total_value','qnt_orders']]
```

```python
features
```

We have a big difference between the values, being necessary to apply a regularization# We have a big difference between the values, being necessary to apply a regularization

```python
# Generating the new pattern
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)
```

### Applying K-Means

```python
 from sklearn.cluster import KMeans
```

```python
wcss = []
range_values = range(1, 20)
for i in range_values:
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(features_scaled)
    wcss.append(kmeans.inertia_)
    
    
```

```python
import matplotlib.pyplot as plt
plt.plot(wcss, 'bx-')
plt.xlabel('Clusters')
plt.ylabel('WCSS')
```

using inertia to find the best amount of clusters

```python
kmeans = KMeans(n_clusters=4)
kmeans.fit(features_scaled)
```

```python
customer_info['cluster'] = kmeans.labels_
```

```python
customer_info
```

```python
plt.figure(figsize=(10,7))
ax = sns.scatterplot(x= 'Total_value', y='qnt_orders', hue='cluster',data=customer_info,palette=['red','green','blue','orange'])
plt.show()
```

#### Conclusion

We can see that the Blue cluster has the highest expenditure among the other clusters... The Orange cluster has the highest amount of requests...


With this, it is possible to infer that the customers of the Blue clusters have a higher expense and take less damage to the company (since they place fewer orders but at higher values and has a lower cost of transport). The orange clusters, on the other hand, have a greater possibility of buying products in large quantities, enabling a campaign to sell/test new products.
